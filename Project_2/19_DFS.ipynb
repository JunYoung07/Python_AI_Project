# 2. DFS 알고리즘 구현
* **스택**과 **큐**자료구조 활용
* `need_visit`스택과 `visitied`큐

```python
# DFS 구현하기
def dfs(graph, start_node):
    visited, need_visit = [], []

    need_visit.append(start_node)

    while need_visit:
        node = need_visit.pop()
        if node not in visited:
            visited.append(node)
            need_visit.extend(graph[node])

    return visited
```
