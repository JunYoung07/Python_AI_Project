# 문자열 함수

## 표준 파이썬 함수 VS 문자열 함수
* 표준 파이썬 함수 -> 파이썬이 실행되면 바로 사용할 수 있는 함수
* 문자열 함수 -> 문자열이 선언된 경우에만 사용할 수 있는 함수
* 호출방식이 다르다!!

* 문자열과 관련된 여러가지 기능(함수)


---
```
# count(): 문자열에서 매개변수로 전달된 문자여르이 출현 횟수를 반환
# find(): 매개변수로 전달된 문자열이 처음 나오는 위치를 인덱스로 반환 (없으면 -1 반환)
# index(): 매개변수로 전달된 문자열이 처음 나오는 위치를 인덱스로 반환 (없으면 Error)
# replace(): 특정 문자열을 치환
# upper(): 문자열을 모두 대문자로 변환
# lower(): 문자열을 모두 소문자로 변환
# lstrip(): 문자열의 왼쪽 공백을 제거
# rstrip(): 문자열의 오른쪽 공백을 제거
# strip(): 문자열의 양쪽 공백을 제거
# join(): 문자열에 특정 문자열을 삽입
```





````python
# class str -> count()
str = '문자열 함수는 문자열과 관련된 여러가지 기능'
print(str.count('문'))
````

````python
# class str -> find()
print(str.find('함'))
print(str.find('abc'))
````

````python
# class str -> index()
print(str.index('함'))
print(str.index('abc'))
````

````python
# class str -> replace()
print(str.replace(' ', '/'))
````

````python
# class str -> upper()
# class str -> lower()
str = 'python PYTHON'
print(str.upper())
print(str.lower())
````

````python
# class str -> lstrip()
# class str -> rstrip()
# class str -> strip()
str = '     Hello       Python      '
print(str.lstrip())
print(str.rstrip())
print(str.strip())
````

````python
# class str -> join()
str = 'python'
print(str.join('..'))
print('.'.join(str))
````
