# 2. 스타일 옵션

```python
plt.figure(figsize=(6, 8))
plt.plot([1, 2, 3], [1, 2, 3])
plt.plot([1, 2, 3], [2, 4, 6])

plt.title('타이틀 테스트', fontsize=15)
plt.xlabel('x축', fontsize=15)
plt.ylabel('y축', fontsize=15, rotation=0)

plt.show()
```

```python
plt.figure(figsize=(15, 10))
plt.title('마커 설정', fontsize=20)
plt.plot(np.arange(10), np.arange(10), marker='o', markersize=15, linestyle='', color='b', alpha=0.3)
plt.plot(np.arange(10), np.arange(10)*2, marker='v', markersize=15, linestyle='--', color='r', alpha=0.5)
plt.plot(np.arange(10), np.arange(10)*3, marker='*', markersize=15, linestyle='-.', color = 'y', alpha=1)

# 범례 생성
plt.legend(['10', '10*2', '10*3'], fontsize=15, loc='lower right', ncol=3)

# 축 범위 지정
plt.xlim(0, 12)
plt.ylim(0, 30)

# 라벨의 속성 변경
plt.xticks(rotation=30)
plt.yticks(rotation=90)

# 눈금자 추가
plt.grid()

plt.show()
```

```python
x = ["Math", "Programming", "Data Science", "Art", "English", "Korean"]
y = [60, 80, 50, 90, 20, 70]

plt.figure(figsize=(8, 5))
plt.bar(x, y, align='center', alpha=0.7, color='red')
plt.ylabel("Score of Student")
plt.title("Subjects")
plt.show()
```

```python
x = ["Math", "Programming", "Data Science", "Art", "English", "Korean"]
y = [60, 80, 50, 90, 20, 70]

plt.figure(figsize=(8, 5))
plt.barh(x, y, align='center', alpha=0.7, color='pink')
plt.yticks(x)
plt.xlabel("Score of Student")
plt.title("Subjects")
plt.show()
```
