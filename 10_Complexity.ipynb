````python
# 1ë¶€í„° 100ê¹Œì§€ì˜ í•©ì„ êµ¬í•˜ëŠ” ì•Œê³ ë¦¬ì¦˜
def sum_all(n):
    total = 0
    for num in range(1, n+1):
        total += num
    return total
````
